```python
#Q1
def main():
    for i in range(1, int(input()) + 1):
        D ,P = input().split(" ")
        D= int(D)
        ans =solution(D,P)
        print('Case #{}: {}'.format(i, ans))
    return

def solution(d,p):
    c=0
    dic = {}
    i=0
    for char in p:
        if char == 'S':
            dic[c]= dic.get(c,0)+1
        else:
            c+=1
    score = get_score(dic)
    if p.count('S')>d:
        return 'IMPOSSIBLE'
    swap = 0
    while d-score<0:
        #swap increase
        #update score
        swap +=1
        ind = max(dic.keys())
        dic[ind]-=1
        if dic[ind]==0:
            dic.pop(ind,None)
        dic[ind-1] = dic.get(ind-1,0) +1
        score -= 2**ind-2**(ind-1)        
    return swap
    
            
def get_score(dic):
    total = 0
    for key in dic:
        total+=2**key*dic[key]
    return total

main()
```

```python
#Q2
def main():
    for i in range(1, int(input()) + 1):
        N = int(input())
        ls = [int(s) for s in input().split(" ")]
        ans =solution(ls,N)
        print('Case #{}: {}'.format(i, ans))
        
def solution(ls,N):
    evenls=ls[::2]
    oddls = ls[1::2]
    evenls.sort()
    oddls.sort()
    zip1 = zip(evenls,oddls)
    zip2=zip(oddls,evenls[1:])
    c = None
    for i,(a,b) in enumerate(zip1):
        if a>b:
            c=2*i
            break
    for i,(a,b) in enumerate(zip2):
        if a>b:
            if c is None or 2*i+1<c:
                c=2*i+1
            break
    if c is None:
        return 'OK'
    return c
    
main()
```

```python
#Q4
from math import pi,cos,sin,asin,acos,tan,atan,sqrt

def main():
    for i in range(1, int(input()) + 1):
        N = float(input())
        print('Case #{}:'.format(i))
        solution(N)
    return

def rotate_z(c0,angle = pi/4):
    x,y,z=c0
    return (cos(angle)*x-y*sin(angle),x*sin(angle)+y*cos(angle),z)

def rotate_x(c,tetha = pi/4):
    x,y,z=c
    return (x,y*cos(tetha)-z*sin(tetha),y*sin(tetha)+z*cos(tetha))

def rotate_y(c,tetha = pi/4):
    x,y,z = c
    return x*cos(tetha)+z*sin(tetha),y,-x*sin(tetha)+z*cos(tetha)

def solution(N):
    x1 = (0.5,0.5,-0.5)
    x2 = (-0.5,-0.5,0.5)
    x1_new = rotate_y(x1)
    x2_new = rotate_y(x2)
    #print(x1_new,x2_new)
    a = x2_new[2]-x1_new[2]
    b = x1_new[1]-x2_new[1]
    R = sqrt(a**2+b**2)
    alpha = atan(b/a)
    tetha = asin(N/R)-alpha
    f1 = (0,0,0.5)
    f2= (0,0.5,0)
    f3 = (0.5,0,0)
    f1 = rotate_x(rotate_y(f1),tetha)
    f2 = rotate_x(rotate_y(f2),tetha)
    f3 = rotate_x(rotate_y(f3),tetha)
    print('{} {} {}'.format(*f1))
    print('{} {} {}'.format(*f2))
    print('{} {} {}'.format(*f3))
    
main()
```
