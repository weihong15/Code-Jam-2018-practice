```python
#alantian

#!/usr/bin/env python3

import sys
from collections import Counter

def read_str():
    return input()

def read_strs():
    return list(input().split())

def read_ints():
    return list(map(int, input().split()))

def solve(case_no):
    R, B, C = read_ints()

    cashiers = [
        read_ints()
        for i in range(C)
    ]

    M, S, P = list(zip(*cashiers))

    def ok(max_time):
        can_process = [
            min(max(0, max_time - P[i]) // S[i], M[i])
            for i in range(C)
        ]
        # print('ok', max_time, can_process)
        can_process.sort()
        return sum(can_process[-R:]) >= B

    left = 0
    right = max(S) * B + max(P)

    while left < right:
        mid = (left + right ) // 2
        if not ok(mid):
            left = mid + 1
        else:
            right = mid

    ans = left

    # output answer
    print('Case #%d: %s' % (case_no, ans))

def main():
    T, = read_ints()
    for t in range(T):
        solve(t+1)

if __name__ == '__main__':
    main()
```
