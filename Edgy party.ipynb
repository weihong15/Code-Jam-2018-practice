```python
#alantian

#!/usr/bin/env python3

import sys
from collections import Counter
import math

def read_str():
    return input()

def read_strs():
    return list(input().split())

def read_ints():
    return list(map(int, input().split()))

def read_floats():
    return list(map(float, input().split()))

def new_ranges(ranges, a, b):
    new_ranges = list(ranges)
    for _a, _b in ranges:
        new_ranges.append((_a + a, _b + b))

    events = []
    for a, b in new_ranges:
        events.append((a, -1))  # stack in
        events.append((b, +1))  # stack out
    events.sort()

    final_ranges = []
    #print('events', events)
    acc = 0
    last_pos_0 = 0.
    for pos, d in events:
        acc += d
        #print('acc, d', acc, d)
        if acc == 0 and d == 1:
            final_ranges.append((last_pos_0, pos))
        elif acc == -1 and d == -1:
            last_pos_0 = pos

    return final_ranges

def solve(case_no):
    N, P = read_ints()
    P = float(P)

    cookies = [read_floats() for i in range(N)]

    # W, H = list(zip(*cookies))

    for W, H in cookies:
        P -= (W + H) * 2

    #print('P', P)

    def tri(a, b):
        return math.sqrt(a*a + b*b)

    ranges = [[0., 0.]]


    for W, H in cookies:
        this_a = min(W, H) * 2.0
        this_b = tri(W, H) * 2.0
        _ranges = new_ranges(ranges, this_a, this_b)
        ranges = _ranges
        #print(H, W, ranges)


    best_v = 0.
    for a, b in ranges:
        if a <= P <= b:
            best_v = max(best_v, P)
        elif b < P:
            best_v = max(best_v, b)

    for W, H in cookies:
        best_v += (W + H) * 2

    ans = best_v

    # output answer
    print('Case #%d: %.10f' % (case_no, ans))

def main():
    T, = read_ints()
    for t in range(T):
        solve(t+1)

if __name__ == '__main__':
    main()
```

```python
#mcpower

import math
from pprint import pprint
def get_range(W, H):
	small = min(W, H)
	large = (W**2 + H**2) ** 0.5
	return [2 * small, 2 * large]

def get_perim(W, H):
	return 2 * (W + H)

MAX_SIZE = math.ceil( (250 * 4 + 250 * 2 * math.sqrt(2))* 100 )
# lol
MAX_SIZE += 100

def do_case():
	# input()
	# int(input())
	N, P = list(map(int, input().split()))

	P = min(P, MAX_SIZE)
	
	cookies = []
	for i in range(N):
		W, H = list(map(int, input().split()))
		cookies.append([W, H])

	# P is at least as large as sum of perim
	orig_perim = sum(get_perim(W, H) for W, H in cookies)
	target = P - orig_perim
	# print(P, orig_perim, target)

	ranges = [get_range(W, H) for W, H in cookies]

	# sum of max < target? return that.
	max_shot = sum(large for small, large in ranges)
	if max_shot <= target:
		return orig_perim + max_shot

	# we know that we can definitely EXCEED the target
	min_shot_all = sum(small for small, large in ranges)
	if min_shot_all <= target:
		# that means we can always get it
		return P

	# we know that we exceed the target in both min shot and max shot.

	# now we can't take everything

	# do a DP!
	# return fudge
	dp = [[None] * N for _ in range(target + 10)]

	
	# return fudge factor
	# do this over all p, 0 from target downwards
	def recurse(p, ind):
		# print(p, ind)
		# if p < 0:
		# 	return -MAX_SIZE
		if ind == N:
			if p == 0:
				return 0
			return -MAX_SIZE
		if dp[p][ind] is not None:
			return dp[p][ind]

		# don't take it
		out = recurse(p, ind + 1)

		# take it
		small, large = ranges[ind]
		if p - small >= 0:
			out = max(out, recurse(p - small, ind + 1) + large - small)

		dp[p][ind] = out

		return out

	# to target
	out = 0
	# print(ranges)
	for p in range(0, target+1):
		o = recurse(p, 0)
		# print(p, o)

		if o >= 0:
			out = max(out, min(target, p + o))

	# pprint(dp)
	return orig_perim + out
	

	
	

T = int(input())
for case in range(T):
	ans = do_case()
	print("Case #{}: {}".format(case+1, ans))
```

```python
def solve(N, L, C):
    left = N - sum(C) #people voted left
    C.extend([0] * left) #extend till L + left bins
    mods = [x % N for x in C] #remainder
    vals = C[:] #=C make a copy, won't edit original array
    
    fracs = [((100 * i) / N) % 1 >= 0.5 for i in range(N)] #which is round up
    dist = [100000000000 for i in range(N)] 

    min_true = None
    for i in range(N):
        if fracs[i]:
            min_true = i + N
            break
    #frac()
    
    
    if min_true is not None:
        for i in range(N - 1, -1, -1):
            if fracs[i]:
                min_true = i
            dist[i] = min_true - i
    # print(dist)

    dd = sorted([(dist[x], i) for (i, x) in enumerate(mods)])


    max_i = 0
    left_used = left
    while max_i < len(dd) and left_used >= dd[max_i][0]:
        left_used -= dd[max_i][0]
        max_i += 1

    for i in range(max_i):
        vals[dd[i][1]] += dd[i][0]
    vals[0] += left_used
    
    rounded = [get_round((100 * v), N) for v in vals]
    # print(vals, [round(100 * v / N, 3) for v in vals])
    return sum(rounded)

def get_round(n, d):
    q = n // d
    rem = n - q * d
    if rem * 2 >= d:
        return q + 1
    return q


N = int(input())
for case in range(N):
    N, L = map(int, input().strip().split())
    C = list(map(int, input().strip().split()))
    print('Case #{}: {}'.format(case + 1, solve(N, L, C)))
```
